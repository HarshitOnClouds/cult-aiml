# Phase 2: Exploratory Data Analysis (EDA)

In this notebook, we analyze the Netflix Prize Dataset to extract business insights and understand the data distribution.
We will load the preprocessed `tiny_subset.parquet` (or `medium_subset`) to perform this analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid")

# Load the dataset
data_path = os.path.join('..', 'data', 'subsets', 'tiny_subset.parquet')
if not os.path.exists(data_path):
    print("Subset not found. Please run the data pipeline first.")
else:
    df = pd.read_parquet(data_path)
    print(f"Data loaded! Shape: {df.shape}")
    display(df.head())
```

## 1. Rating Distribution Analysis

```python
if 'df' in locals():
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x='rating', palette='viridis')
    plt.title('Overall Rating Distribution')
    plt.xlabel('Rating (Stars)')
    plt.ylabel('Count')
    plt.show()
    
    # Check for 4-star bias: Is 4 the most common rating?
```

```python
if 'df' in locals():
    # Per-user average rating
    user_avg = df.groupby('user_id')['rating'].mean()
    plt.figure(figsize=(8, 5))
    sns.histplot(user_avg, bins=30, kde=True, color='skyblue')
    plt.title('Distribution of Average Rating per User')
    plt.xlabel('Average Rating')
    plt.show()
```

## 2. Activity Analysis (Power Law)

```python
if 'df' in locals():
    user_activity = df['user_id'].value_counts()
    movie_activity = df['movie_id'].value_counts()
    
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))
    
    sns.histplot(user_activity, bins=50, log_scale=(False, True), ax=axes[0], color='coral')
    axes[0].set_title('Ratings per User (Log Scale)')
    axes[0].set_xlabel('Number of Ratings')
    axes[0].set_ylabel('Number of Users')
    
    sns.histplot(movie_activity, bins=50, log_scale=(False, True), ax=axes[1], color='mediumseagreen')
    axes[1].set_title('Ratings per Movie (Log Scale)')
    axes[1].set_xlabel('Number of Ratings')
    axes[1].set_ylabel('Number of Movies')
    
    plt.tight_layout()
    plt.show()
```

## 3. Sparsity Analysis

```python
if 'df' in locals():
    n_users = df['user_id'].nunique()
    n_movies = df['movie_id'].nunique()
    n_ratings = len(df)
    
    sparsity = 1.0 - (n_ratings / (n_users * n_movies))
    print(f"Total Users: {n_users:,}")
    print(f"Total Movies: {n_movies:,}")
    print(f"Total Ratings: {n_ratings:,}")
    print(f"Matrix Sparsity: {sparsity * 100:.4f}%")
```

## 4. Business Insights & Summary

- **Long Tail:** A tiny fraction of movies account for the vast majority of ratings.
- **Cold Start:** Even after sampling, many users have very few ratings, highlighting the need for cold-start strategies.
- **Rating Bias:** Users tend to give 3 or 4-star ratings much more frequently than 1 or 2 stars.
